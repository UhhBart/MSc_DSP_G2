```python
import osmnx as ox

def get_pois(city, poi_type):
    """
    Fetch points of interest of a specific type in a given city.

    Args:
    city (str): Name of the city.
    poi_type (str): Type of point of interest.

    Returns:
    GeoDataFrame: A GeoDataFrame containing points of interest.
    """
    # Define the query for the specified city
    place_query = {"city": city, "country": "Netherlands"}
    
    # Retrieve POIs
    pois = ox.features_from_place(place_query, tags={poi_type: True})
    
    return pois

# Example usage
city_name = "Amsterdam"
poi_type = "amenity"

amenities = get_pois(city_name, poi_type)

display(amenities)
```

```python
import geopandas as gpd
import pandas as pd
from scipy.spatial import cKDTree
from shapely.wkt import loads

# Correct CRS for Amsterdam
crs_amsterdam = "EPSG:28992"


prio_pois_df = amenities[amenities.amenity.isin(prio_prio_pois)]
grid_df = pd.read_csv("final_data/grids/grid_by_hand.csv", sep=",", encoding="utf-8")

# Ensure geometry is correctly loaded
grid_df['geometry'] = grid_df['geometry'].apply(loads)
grid_gdf = gpd.GeoDataFrame(grid_df, geometry='geometry', crs=crs_amsterdam)
prio_pois_df = prio_pois_df.set_crs(crs_amsterdam, allow_override=True)

# Convert all data to the same CRS (EPSG:28992)
grid_gdf = grid_gdf.to_crs(crs_amsterdam)
prio_pois_df = prio_pois_df.to_crs(crs_amsterdam)

# Step 2: Create buffers around grid cells
meters = 500
buffer_num = meters * (1 / 68000.09)
grid_gdf['buffer'] = grid_gdf.geometry.buffer(buffer_num, resolution=16)  

# Check for valid geometries in grid_gdf and prio_pois_df
if not grid_gdf.geometry.is_valid.all() or not prio_pois_df.geometry.is_valid.all():
    raise ValueError("Invalid geometries found in one or more GeoDataFrames.")

# Step 3: Build a KD-tree from POI coordinates
poi_coordinates = prio_pois_df.geometry.apply(lambda geom: (geom.centroid.x, geom.centroid.y)).tolist()
poi_tree = cKDTree(poi_coordinates)

def calculate_distances_within_buffer(row):
    buffer = row.buffer
    nearby_pois = prio_pois_df[prio_pois_df.geometry.within(buffer)]

    if len(nearby_pois) > 0:
        poi_coordinates = nearby_pois.geometry.apply(lambda geom: (geom.centroid.x, geom.centroid.y)).tolist()
        poi_tree = cKDTree(poi_coordinates)
        grid_centroid = row.geometry.centroid
        distances, poi_indices = poi_tree.query((grid_centroid.x, grid_centroid.y), k=len(nearby_pois))
        return distances, poi_indices
    else:
        return [], []

# Apply the distance calculation function to each row in grid_gdf
grid_gdf[['distances', 'poi_indices']] = grid_gdf.apply(calculate_distances_within_buffer, axis=1, result_type='expand')

display(grid_gdf)
```
